```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = "spojeny.csv" 
df = pd.read_csv(file_path)

# styl pre grafy
sns.set_style("whitegrid")

# Rozdelenie atributov na numericke a kategoricke
numeric=df[['caseend', 'opstart', 'opend', 'case_duration',  'op_duration', 'age', 'height', 'weight', 'bmi', 'icu_days',  'preop_hb', 'preop_plt', 'preop_pt', 'preop_aptt', 'preop_na', 'preop_k', 'preop_gluc', 'preop_alb', 'preop_ast', 'preop_bun', 'preop_ph', 'preop_hco3', 'preop_be', 'preop_pao2', 'preop_paco2', 'preop_sao2', 'intraop_ebl', 'intraop_uo',  'intraop_crystalloid',  'intraop_ppf',   'intraop_rocu',  'intraop_eph',  'hospital_stay']]
nominal=df[['sex','department','approach','ane_type','death_inhosp','cormack','airway','tubesize','dltubesize','lmasize','aline2','emop','hypotenzia','asa','preop_htn','preop_dm']]

for col in numeric.columns:
    plt.figure(figsize=(8, 4))
    # palette = sns.color_palette("tab10")
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot graf pre: {col}')
    # plt.savefig(f"boxplot_{col}.png", dpi=300, bbox_inches='tight')  
    plt.show()

for col in nominal.columns:
    plt.figure(figsize=(8, 4))
    unique_vals = df[col].value_counts().index
    palette = sns.color_palette("tab10", len(unique_vals))
    sns.countplot(
        data=df,
        x=col,
        hue=col,
        order=unique_vals,
        palette=palette,
        legend=False,
        dodge=False
    )
    
    plt.title(f'Barplot pre: {col}')
    plt.xlabel(col)
    plt.ylabel('Počet výskytov')
    # plt.savefig(f"barplot_{col}.png", dpi=300, bbox_inches='tight')  
    plt.show()
    
```

```python


existing_attributes=df[['intraop_ca', 'intraop_colloid', 'intraop_epi', 'intraop_ffp', 'intraop_ftn' , 'intraop_mdz', 'intraop_phe', 'intraop_rbc', 'intraop_vecu']]
tab10_colors = plt.get_cmap("tab10", len(existing_attributes.columns))
for i, attr in enumerate(existing_attributes):
    plt.figure()
    categorized = df[attr].apply(lambda x: "Podané" if x > 0 else "Nepodané")
    value_counts = categorized.value_counts()
    colors = [tab10_colors(0), tab10_colors(1)]
    value_counts.plot(kind='bar', color=colors)
    plt.title(f"Podanie: {attr}")
    plt.ylabel("Počet prípadov")
    plt.xticks(rotation=0)
    # plt.savefig(f"plot{attr}.png", dpi=300, bbox_inches='tight')
    plt.tight_layout()
    plt.show()
    
```

```python
col = "anestart"

#  IQR 
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 4 * IQR
upper_bound = Q3 + 1.5 * IQR

# Boxplot s upravenou Y-osou
plt.figure(figsize=(8, 4))
sns.boxplot(y=df[col])

plt.ylim(lower_bound, upper_bound)  # Nastavenie rozsahu na IQR
plt.title(f'Boxplot pre: {col}')
plt.savefig(f"{col}.png", dpi=300, bbox_inches='tight')
plt.ylabel(col)

plt.show()
```

```python
col = "preop_cr"


Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 2 * IQR
upper_bound = Q3 + 6 * IQR


plt.figure(figsize=(8, 4))
sns.boxplot(y=df[col])

plt.ylim(lower_bound, upper_bound) 
plt.title(f'Boxplot pre: {col}')
plt.savefig(f"{col}.png", dpi=300, bbox_inches='tight')
plt.ylabel(col)

plt.show()
```
